## Explore the phone_brand_device_model dataset

```python
# imports
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
%matplotlib inline

import sys
reload(sys)
sys.setdefaultencoding("utf-8")

import matplotlib.font_manager as fm
font = fm.FontProperties(fname='c:\\windows\\fonts\\simsun.ttc')
```

```python
phone = pd.read_csv('phone_brand_device_model.csv', encoding='utf-8')
```

```python
phone.head(10)
```

```python
phone.shape
```

```python
# Are there duplicated device ids?

phone2 = phone.drop_duplicates(['device_id'])
```

```python
phone2.shape
```

There are few rows after dropping duplicates. So there indeed are duplicated device ids.

```python
plot1 = sns.factorplot('phone_brand', data = phone2, kind = 'count', size = 15, aspect = 2)
plot1.set_xticklabels(phone2.phone_brand.str.decode('utf-8'), fontproperties = font)
```

The numbers of phone brand/device model combinations can vary significantly accorss all the brands. 

```python
popular = phone2.phone_brand.value_counts()
popular.sort_values(ascending = False, inplace = True)
```

```python
popular.shape
```

There are 131 distinct brands. 

```python
mpl.rcParams['figure.figsize'] = (8.0, 6.0)
plot2 = popular[0:9].plot(kind = 'bar')
plot2.set_xticklabels(popular.index.str.decode('utf-8'), fontproperties = font)
```

We can zoom in the previous plot to only look at the top 10 brands with the max number of distinct phone brand/device model combinations. 小米，三星，华为 have over twice as many combinations as the rest brands.

```python
least_popular = phone2.phone_brand.value_counts()
least_popular.sort_values(ascending = True, inplace = True)
```

```python
mpl.rcParams['figure.figsize'] = (8.0, 6.0)
plot3 = least_popular[0:9].plot(kind = 'bar')
plot3.set_xticklabels(least_popular.index.str.decode('utf-8'), fontproperties = font)
plot3.set(ylim = (0, least_popular[9] + 1))
```

The 10 least surveyed phone brands only have 1 associated device model.
